```python
import pandas as pd  
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# reading in the data to a dataframe
penglings_df = pd.read_csv('penglings.csv')

colors = {'Adelie': 'darkorange', 'Gentoo': '#008B8B', 'Chinstrap': 'darkorchid'}

plt.scatter(
    x=penglings_df['flipper_length_mm'],
    y=penglings_df['body_mass_g'],
    s=penglings_df['bill_length_mm'],
    c=penglings_df['species'].map(colors)
)

plt.legend(colors.values())
```
